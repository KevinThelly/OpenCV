```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank_img=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,(300,10),(500,100),color=(0,0,255),thickness=(10))
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(255,0,0),thickness=(10))
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img,center=(100,100),radius=(100),color=(0,255,0),thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(0,500),pt2=(500,500),color=(255,255,0),thickness=8)
plt.imshow(blank_img)
```

```python
font= cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank_img,"Kevin",(10,490), font,2,color=(255,0,9),thickness=2)
plt.imshow(blank_img)
```

```python
vertices=np.array([[0,500],[500,0],[50,50],[430,90]])
```

```python
#use the same parameters to change a 2 dimension to 3 dimension for gettig thecolour channels 
pts=vertices.reshape((-1,1,2))
pts
```

```python
blank_img = np.zeros((512,512,3))
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,255,255),thickness=5)
plt.imshow(blank_img)
```
