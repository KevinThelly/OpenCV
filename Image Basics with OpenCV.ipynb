```python
# matplotlib --> RGB RED GREEN BLUE
# opencv ---> BLUE GREEN RED
#so the below image is shown in different colour than original
plt.imshow(img)
```

```python
#image fixed
plt.imshow(fix_img)
```

```python
plt.imshow(img_gray)
#doesnt show gray cos of the working of the imshow
#imshow shows darker shades as blue and lighter in dfferent color
```
